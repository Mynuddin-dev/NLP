- TF (Term Frequency) = Number of Repetation of word in a Sentence / Total word in a Sentence
- IDF (Inverse Data Frequency (IDF)) = Log * (Total Number Of Sentencesss / Number of Sentencesss Contain the Words)
- TF-IDF = TF * IDF

## Text Cleaning Part

```python
# Text Cleaning
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
# ps = PorterStemmer()  not using this because here Lemmatization is better than stemming
nltk.download('punkt')
wordnet=WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
# ps = PorterStemmer()  not using this because here Lemmatization is better than stemming
nltk.download('punkt')
wordnet=WordNetLemmatizer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
sentences_final = []

for i in range(len(sentences)):
    sen = re.sub('[^a-zA-Z]', ' ', sentences[i])  # Replace punctuation with space.
    sen = sen.lower()   # Lowercase
    sen_words = sen.split()   # Tokenization
    sen_final_words = [wordnet.lemmatize(word) for word in sen_words if not word in set(stopwords.words('english'))]   # Skip or Remove Stop words and lemmatize

    final_sen = ' '.join(sen_final_words)
    
    sentences_final.append(final_sen)


# Every cell contains a number, that represents the count of the word in that particular text.
# All words have been converted to lowercase.
# The words in columns have been arranged alphabetically
```

## TF-IDF vectorizer Vs Count Vectorizer (BoW)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
countvectorizer = CountVectorizer()
tfidfvectorizer = TfidfVectorizer()
```

```python
count_wm = countvectorizer.fit_transform(sentences_final)
tfidf_wm = tfidfvectorizer.fit_transform(sentences_final)
```

```python
count_tokens = countvectorizer.get_feature_names()
tfidf_tokens = tfidfvectorizer.get_feature_names()
print(count_tokens)
print(tfidf_tokens)
```
